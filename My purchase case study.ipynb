# STATISTICAL ANALYSIS

### One Sample Test For MeanÂ¶
Let's Suppose, it was observed that the average purchase made by the Men of the age 18-25 was 10000. Is it still the same?



```python
#null hypothesis : the mean is 10000
#alternate hypothesis : the mean is 1000 
```

```python
data['Age'].unique()
```

```python
data['Gender'].unique()
```

```python
new_data=data.loc[(data['Age']==1) & (data['Gender']==1)]
new_data
```

```python
new_data.shape
```

```python
sample_size=1000
sample=new_data.sample(sample_size,random_state=0)
sample
```

```python
new_data.mean()
```

```python
pur_popmean=9440.512276
```

```python
pur_sampmean=sample['Purchase'].mean()
pur_sampmean
```

```python
#one_sample test
from scipy.stats import ttest_1samp
t_stat,p_value=ttest_1samp(sample['Purchase'],pur_popmean)
print('t_stat: ',t_stat)
print('p_value: ',p_value)
```

#### P-value is greater than 0.05, so we accept the null hypothesis. 
#### Therefore, the mean purchase for men aged 18-25 is 10000 and it is still the same.

### One Sample Test for Proportion
 Let's Suppose, It was observed that the percentage of women that spend more than 10000 was 35%. Is it still the same?

```python
#Null hypoyhesis: proportion is 0.35
#Alternate hypothesis: proportion is not equal to 0.35
```

```python
new_data2=data.loc[data['Purchase']>10000]
new_data2
```

```python
new_data2.shape
```

```python
new_data2['Gender'].value_counts()
```

```python
#No of women in the sample
num_fem=new_data2['Gender'].value_counts()[0]
num_fem
```

```python
#number of observations
no_obs=len(new_data2['Gender'])
no_obs
```

```python
#hypothesised value
p0 = 0.35
```

```python
new_data2['Gender'].value_counts()[0]/no_obs
```

```python
#one sample z test
from statsmodels.stats.proportion import proportions_ztest
z_stat,p_value=proportions_ztest(count=num_fem,nobs=no_obs,value=p0,alternative='two-sided',prop_var=False)
print('z_stat: ',z_stat)
print('p_value: ',p_value)
```

#### P-value is less than 0.05, so we reject the null hypothesis.
#### That is, the percentage of women that spend more than 10000 is not 35%

## Two Sample test for Means
Is the average purchase made by men and women of the age 18-25 same?

```python
#null hypothesis - average purchase made are equal 
#alternate hypothesis - average purchase made are not equal
```

```python
data_male=data.loc[(data['Gender']==1) & (data['Age']==1)]
data_female=data.loc[(data['Gender']==0) & (data['Age']==1)]
```

```python
#sample variance
print('data_male(variance): ',data_male['Purchase'].var())
print('data_female(variance): ',data_female['Purchase'].var())
```

```python
#sample mean
print('data_male(mean): ',data_male['Purchase'].mean())
print('data_female(mean): ',data_female['Purchase'].mean())
```

```python
#two sample t-test for unequal variances
from scipy.stats import ttest_ind

t_stat,p_value=ttest_ind(data_male['Purchase'],data_female['Purchase'],equal_var=False)
print('t_stat : ',t_stat)
print('p_value : ',p_value)
```

#### P-value is greater than 0.05, so we accept null hypothesis.
#### That is, the average purchase made by men and women of the age 18-25 are same

## Two Sample test for ProportionÂ¶
Is the percentage of men who have spend more than 10000 same for the ages 18-25 and 26-35

```python
#null hypothesis - Percentage of the men in the age groups is same
#alternate hypothesis - Percentage of the men in the age groups is not same
```

```python
data_age1 = data.loc[(data['Age'] == 1) & (data['Purchase'] > 10000)]
data_age2 = data.loc[(data['Age'] == 2) & (data['Purchase'] > 10000)]
```

```python
data_age1.shape
```

```python
data_age2.shape
```

```python
len(data_age1)
```

```python
len(data_age2)
```

```python
data_age1_samp=data_age1.sample(1000,random_state=0)
data_age2_samp=data_age2.sample(1000,random_state=0)
```

```python
count = [(data_age1_samp['Gender'] == 1).sum(), (data_age2_samp['Gender'] == 1).sum()]
print(count)
no_obs = [(len(data_age1_samp)), len(data_age2_samp)]
print(no_obs)
```

```python
from statsmodels.stats.proportion import proportions_ztest
stat_2sample,p_value=proportions_ztest(count=count,nobs=no_obs,value=0,alternative='two-sided',prop_var=False)
print(stat_2sample,p_value)
```

#### P value is more than 0.05, so accept the null hypothesis. 

#### Therefore, Percentage of the men in  both the age groups who have spent more than 10000 is same
